```python
def get_age_canadian_classif(x):
  if not type(x) == int or type(x) == float:
    return 'none'
  else:
    if x in range(25, 65):
      return 'adult'
    elif x in range(15, 25):
      return 'youth'
    elif x > 64:
      return 'senior'
```

```python
df.age.apply(get_age_canadian_classif)
```

```python
df['age_can_clasif'] = df.age.apply(get_age_canadian_classif)
```
